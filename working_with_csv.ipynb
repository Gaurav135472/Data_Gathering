# Convertors 

```python
def rename(data):
    if data == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return data
    
pd.read_csv('IPL Matches 2008-2020.csv', converters={'team1' : rename}).head()
```

# na_values param

```python
pd.read_csv('aug_train.csv', na_values=['Male']).head()
```
